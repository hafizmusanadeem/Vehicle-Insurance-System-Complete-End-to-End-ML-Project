# **Scope of this Notebook is to only push a local data to MongoDB so then we can pull our data from an actual database.**

```python
import pandas as pd
import pymongo

df = pd.read_csv('data.csv')
```

```python
# data should be converted into dictiionary before we push it to MongoDB
data = df.to_dict(orient= 'records')
```

```python
DB_NAME = 'Proj1'
COLLECTION_NAME = 'Proj1-Data'
CONNECTION_URL = "YOUR_CONNECTION_URL"
```

```python
client = pymongo.MongoClient(CONNECTION_URL)
data_base = client[DB_NAME]
collection = data_base[COLLECTION_NAME]
```

```python
rec = collection.insert_many(data)
```
